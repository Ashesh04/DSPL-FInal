```python
data_path = '/content/drive/MyDrive/demographics_residing_lka.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
# Instead of trying to drop by a specific numerical index,
# reset the index and then drop the first row if needed.
df = df.reset_index(drop=True)  # Reset the index
df = df.drop(index=0)  # Drop the first row (now with index 0)
df.head()
```

```python
df.columns
```

```python
df.drop(['Country of Origin Code', 'Country of Asylum Code'], axis=1, inplace=True, errors='ignore')
df.columns
```

```python
df.info()
```

```python
# List of columns to convert
columns_to_convert = ['Female 0-4', 'Female 5-11', 'Female 12-17', 'Female 18-59',
       'Female 60 or more', 'Female Unknown', 'Female Total', 'Male 0-4',
       'Male 5-11', 'Male 12-17', 'Male 18-59', 'Male 60 or more',
       'Male Unknown', 'Male Total', 'Total']

# Convert columns to numeric, handling errors
for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64') # coerce invalid values to NaN, then convert to Int64
```

```python
df.info()
```

```python
# Check for missing values in the entire DataFrame
missing_values1 = df.isnull().sum()

# Display the count of missing values for each column
print(missing_values1)
```

```python
# Drop rows with any missing values
df.dropna(inplace=True)
missing_vsalues2 = df.isnull().sum()
print(missing_vsalues2)
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```
